```python
#Importing the Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#Data Collection and Preprocessing
#loading the data from sklearn
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
#loading data to dataframe
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
#adding a traget column to the existing dataset
data_frame['label']=breast_cancer_dataset.target
```

```python
data_frame.shape
```

```python
#getting info about data
data_frame.info()
```

```python
#checking for null values
data_frame.isnull().sum()
```

```python
#statistical info about the data
data_frame.describe()
```

```python
#checking out the distribution of the target value
data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
#seperating features and targets
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
#spliting training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
#standardising the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
```

```python
#building the neural network
#importing the dependencies
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting the layers of the neural network
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#training the neural network model
tothistory=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

```python
#visualising accuracy 
plt.plot(tothistory.history['accuracy'])
plt.plot(tothistory.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(tothistory.history['loss'])
plt.plot(tothistory.history['val_loss'])
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training data','validation data'],loc='upper right')
```

```python
loss,accuracy=model.evaluate(x_test_std,y_test)
print(accuracy)
```

```python
y_pred=model.predict(x_test_std)
```

```python
print(y_pred.shape)
print(y_pred[0])
```

```python
#converting the prediction probability to class labels
y_pred_labels=[np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
